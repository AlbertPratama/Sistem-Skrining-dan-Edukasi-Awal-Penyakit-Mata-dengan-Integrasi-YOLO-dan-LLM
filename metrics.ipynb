```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_losses(data_path):
    """
    Plot training and validation losses.
    Fig 1: Training box, cls, dfl losses over epochs.
    Fig 2: Validation box, cls, dfl losses over epochs.
    """
    # Load the data
    df = pd.read_csv(data_path)

    # Figure 1: Training losses
    plt.figure(figsize=(8, 5))
    plt.plot(df['epoch'], df['train/box_loss'], label='Train Box Loss')
    plt.plot(df['epoch'], df['train/cls_loss'], label='Train Cls Loss')
    plt.plot(df['epoch'], df['train/dfl_loss'], label='Train DFL Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss Curves')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Figure 2: Validation losses
    plt.figure(figsize=(8, 5))
    plt.plot(df['epoch'], df['val/box_loss'], label='Val Box Loss')
    plt.plot(df['epoch'], df['val/cls_loss'], label='Val Cls Loss')
    plt.plot(df['epoch'], df['val/dfl_loss'], label='Val DFL Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Validation Loss Curves')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_map_metrics(data_path):
    """
    Plot mAP metrics over epochs.
    Fig 1: metrics/mAP50(B)
    Fig 2: metrics/mAP50-95(B)
    """
    df = pd.read_csv(data_path)

    # Figure 1: mAP@50
    plt.figure(figsize=(8, 5))
    plt.plot(df['epoch'], df['metrics/\u006dAP50(B)'], label='mAP@50(B)')
    plt.xlabel('Epoch')
    plt.ylabel('mAP@50(B)')
    plt.title('mAP@50 over Epochs')
    plt.legend()
    plt.grid(False)
    plt.tight_layout()
    plt.show()

    # Figure 2: mAP@50-95
    plt.figure(figsize=(8, 5))
    plt.plot(df['epoch'], df['metrics/\u006dAP50-95(B)'], label='mAP@50-95(B)')
    plt.xlabel('Epoch')
    plt.ylabel('mAP@50-95(B)')
    plt.title('mAP@50-95 over Epochs')
    plt.legend()
    plt.grid(False)
    plt.tight_layout()
    plt.show()


def plot_precision_recall(data_path):
    """
    Plot precision and recall side by side.
    Left: metrics/precision(B)
    Right: metrics/recall(B)
    """
    df = pd.read_csv(data_path)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Precision plot
    ax1.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Precision')
    ax1.set_title('Precision over Epochs')
    ax1.grid(False)

    # Recall plot
    ax2.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Recall')
    ax2.set_title('Recall over Epochs')
    ax2.grid(False)

    plt.tight_layout()
    plt.show()


if __name__ == '__main__':
    data_file = 'runs/detect/train/results.csv'
    plot_losses(data_file)
    plot_map_metrics(data_file)
    plot_precision_recall(data_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Setting global style untuk publikasi ilmiah
plt.rcParams.update({
    'figure.dpi': 300,
    'font.family': 'serif',
    'font.serif': ['Times New Roman'],
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'legend.fontsize': 10,
    'axes.linewidth': 1.0,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})

def plot_validation_vs_training(data_path):
    df = pd.read_csv(data_path)
    epochs = df['epoch']

    fig, ax = plt.subplots(figsize=(6, 4))
    # Training Loss
    ax.plot(epochs, df['train/box_loss'],  label='Train Box Loss',  linewidth=1.5)
    ax.plot(epochs, df['train/cls_loss'],  label='Train Cls Loss',  linewidth=1.5)
    ax.plot(epochs, df['train/dfl_loss'],  label='Train DFL Loss',  linewidth=1.5)
    # Validation Loss
    ax.plot(epochs, df['val/box_loss'],    linestyle='--', label='Val Box Loss',   linewidth=1.5)
    ax.plot(epochs, df['val/cls_loss'],    linestyle='--', label='Val Cls Loss',   linewidth=1.5)
    ax.plot(epochs, df['val/dfl_loss'],    linestyle='--', label='Val DFL Loss',   linewidth=1.5)

    # Styling
    ax.set_title('Loss Curves')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.grid(which='major', linestyle=':', linewidth=0.8)
    # Hilangkan spines atas & kanan
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Legend di luar plot
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', frameon=True)
    plt.tight_layout()
    plt.show()
```
